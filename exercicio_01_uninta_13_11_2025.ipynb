```python
class Pessoa:
    def __init__(self, nome: str, idade: int, peso: float, altura: float):
        """
        altura em cm, peso em kg
        """
        self.nome = nome
        self.idade = idade
        self.peso = peso
        self.altura = altura

    def envelhecer(self, anos: int = 1):
        """
        Envelhece a pessoa 'anos' anos.
        Para cada ano, se a idade for menor que 21, cresce 0,5 cm.
        """
        for _ in range(anos):
            if self.idade < 21:
                self.altura += 0.5   # cresce 0,5 cm
            self.idade += 1

    def engordar(self, kg: float):
        """Aumenta o peso em 'kg' quilogramas."""
        self.peso += kg

    def emagrecer(self, kg: float):
        """Diminui o peso em 'kg' quilogramas (sem deixar negativo)."""
        self.peso -= kg
        if self.peso < 0:
            self.peso = 0

    def crescer(self, cm: float):
        """Aumenta a altura em 'cm' centímetros."""
        self.altura += cm

    def __str__(self):
        return (f"Nome: {self.nome}\n"
                f"Idade: {self.idade} anos\n"
                f"Peso: {self.peso:.1f} kg\n"
                f"Altura: {self.altura:.1f} cm")


# Exemplo de uso
if __name__ == "__main__":
    p = Pessoa("Ana", 18, 55.0, 165.0)

    print("Estado inicial:")
    print(p)

    # Ana envelhece 5 anos
    p.envelhecer(5)        # dos 18 aos 21 ela cresce 0,5 cm/ano; dos 21 aos 23 não cresce

    # Ana engorda 3 kg
    p.engordar(3)

    # Ana emagrece 1,5 kg
    p.emagrecer(1.5)

    # Crescer "manual" (por exemplo, medição errada, ajuste etc.)
    p.crescer(0.5)

    print("\nDepois das alterações:")
    print(p)
```

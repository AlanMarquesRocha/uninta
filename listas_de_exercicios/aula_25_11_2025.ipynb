**1.	Escreva um programa para ler um valor do teclado e apresentar o seu antecessor.**

**3.	Escreva um programa para ler o nmero total de eleitores de um municpio, o nmero de votos brancos, nulos e vlidos. Apresente a percentagem que cada um representa em relao ao total de eleitores.**

```python
# Programa para calcular a porcentagem de votos em uma eleio

# L o nmero total de eleitores
total_eleitores = int(input("Digite o nmero total de eleitores: "))

# L o nmero de votos brancos
votos_brancos = int(input("Digite o nmero de votos brancos: "))

# L o nmero de votos nulos
votos_nulos = int(input("Digite o nmero de votos nulos: "))

# L o nmero de votos vlidos
votos_validos = int(input("Digite o nmero de votos vlidos: "))

# Calcula as porcentagens (regra de trs simples)
porcentagem_brancos = (votos_brancos / total_eleitores) * 100
porcentagem_nulos = (votos_nulos / total_eleitores) * 100
porcentagem_validos = (votos_validos / total_eleitores) * 100

# Exibe os resultados
print("Percentual de votos brancos:", porcentagem_brancos, "%")
print("Percentual de votos nulos:", porcentagem_nulos, "%")
print("Percentual de votos vlidos:", porcentagem_validos, "%")
```

**4.	Escreva um programa que a partir da idade de uma pessoa expressa em anos, meses e dias, apresente a idade apenas em dias (considerar o ano com 365 e cada ms com 30 dias).**



```python
# Programa para converter idade (anos, meses e dias) em apenas dias

# L a idade em anos
anos = int(input("Digite sua idade em anos: "))

# L os meses adicionais
meses = int(input("Digite os meses adicionais: "))

# L os dias adicionais
dias = int(input("Digite os dias adicionais: "))

# Considera 1 ano = 365 dias e 1 ms = 30 dias
total_dias = (anos * 365) + (meses * 30) + dias

# Exibe o total de dias
print("Sua idade em dias :", total_dias, "dias")
```

**6.	Escreva um programa que ajude um comerciante a calcular o valor de venda a partir de um valor de compra de um dado produto.**

image.png

```python
# Programa para calcular o valor de venda com base no valor de compra

# L o valor de compra
valor_compra = float(input("Digite o valor de compra do produto: "))

# Define o percentual de lucro conforme a tabela
if valor_compra < 10:
    lucro = 0.70  # 70%
elif valor_compra < 30:
    lucro = 0.50  # 50%
elif valor_compra < 50:
    lucro = 0.40  # 40%
else:
    lucro = 0.30  # 30%

# Calcula o valor final da venda
valor_venda = valor_compra + (valor_compra * lucro)

# Exibe o resultado
print("Valor de venda do produto: R$", round(valor_venda, 2))
```

**8.	Escreva um programa que leia os limites inferior e superior de um intervalo e imprima todos os nmeros pares no intervalo aberto e o seu somatrio.**

```python
# Programa que imprime os nmeros pares em um intervalo aberto e soma

# L os limites inferior e superior
limite_inferior = int(input("Digite o limite inferior: "))
limite_superior = int(input("Digite o limite superior: "))

soma_pares = 0  # Acumulador

print("Nmeros pares no intervalo aberto:")

# Percorre o intervalo aberto
for numero in range(limite_inferior + 1, limite_superior):
    if numero % 2 == 0:
        print(numero)
        soma_pares += numero

# Exibe a soma dos pares
print("Somatrio dos pares:", soma_pares)
```

**9.	Crie um programa que leia 10 nmeros inteiros e imprima o maior e o menor nmero.**

```python
# Programa que l 10 nmeros e mostra o maior e o menor

# Inicializa as variveis
maior = None
menor = None

for i in range(1, 11):
    numero = int(input(f"Digite o {i} nmero: "))

    # Atualiza maior e menor
    if maior is None or numero > maior:
        maior = numero
    if menor is None or numero < menor:
        menor = numero

# Exibe os resultados
print("Maior nmero:", maior)
print("Menor nmero:", menor)
```

**10.	Escreva um programa que receba vrias idades e que calcule e mostre a mdia das idades. O programa deve finalizar apenas quando for digitada a idade igual a zero.**

```python
# Programa para calcular a mdia de vrias idades at digitar zero

soma_idades = 0
quantidade = 0

while True:
    idade = int(input("Digite uma idade (0 para encerrar): "))

    if idade == 0:
        break

    soma_idades += idade
    quantidade += 1

# Calcula e exibe a mdia, se houve entrada
if quantidade > 0:
    media = soma_idades / quantidade
    print("Mdia das idades:", round(media, 2))
else:
    print("Nenhuma idade foi informada.")
```

**11.	Escreva um programa que permita registar o nome, a altura e o peso de duas pessoas e apresente o nome da mais pesada e o nome da mais alta.**

```python
# Programa que compara altura e peso de duas pessoas

# Pessoa 1
nome1 = input("Nome da primeira pessoa: ")
altura1 = float(input("Altura da primeira pessoa (em metros): "))
peso1 = float(input("Peso da primeira pessoa (em kg): "))

# Pessoa 2
nome2 = input("Nome da segunda pessoa: ")
altura2 = float(input("Altura da segunda pessoa (em metros): "))
peso2 = float(input("Peso da segunda pessoa (em kg): "))

# Verifica o mais alto
if altura1 > altura2:
    print("Pessoa mais alta:", nome1)
elif altura2 > altura1:
    print("Pessoa mais alta:", nome2)
else:
    print("Ambas tm a mesma altura.")

# Verifica o mais pesado
if peso1 > peso2:
    print("Pessoa mais pesada:", nome1)
elif peso2 > peso1:
    print("Pessoa mais pesada:", nome2)
else:
    print("Ambas tm o mesmo peso.")
```

**Utilizando listas faa um programa que faa 5 perguntas para uma pessoa sobre um crime. As perguntas so:**

- a. Telefonou para a vtima?
- b. Esteve no local do crime?
- c. Mora perto da vtima?
- d. Devia para a vtima?
- e. J trabalhou com a vtima?

**O programa deve no final emitir uma classificao sobre a participao da pessoa no crime. Se a pessoa responder positivamente (Sim) a 2 questes ela deve ser classificada como "Suspeita", entre 3 e 4 como "Cmplice" e 5 como "Assassino". Caso contrrio, ele ser classificado como "Inocente".**





```python
# Programa que faz perguntas sobre um crime e classifica a pessoa

# Lista de perguntas
perguntas = [
    "Telefonou para a vtima?",
    "Esteve no local do crime?",
    "Mora perto da vtima?",
    "Devia para a vtima?",
    "J trabalhou com a vtima?"
]

# Contador de respostas positivas
respostas_positivas = 0

# Lao de perguntas
for pergunta in perguntas:
    resposta = input(pergunta + " (sim/no): ").strip().lower()
    if resposta == "sim":
        respostas_positivas += 1

# Classificao com base nas respostas
if respostas_positivas == 2:
    print("Classificao: Suspeita")
elif 3 <= respostas_positivas <= 4:
    print("Classificao: Cmplice")
elif respostas_positivas == 5:
    print("Classificao: Assassino")
else:
    print("Classificao: Inocente")
```
